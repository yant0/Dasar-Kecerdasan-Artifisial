```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
print("Buar graf berarah kosng")
G = nx.DiGraph()
print(f"Graf berarah kosong berhasil dibuat: {G}\n")

plt.figure(figsize=(5,5))
nx.draw(G, with_labels=True, node_color="lightblue", edge_color="gray", arrows=True)
plt.title("Graf Berarah Kosong")
plt.show()
```

Membuat dan menampilkan **graf berarah kosong** menggunakan **NetworkX** dan **Matplotlib**. Objek **DiGraph()** dibuat tanpa simpul atau sisi, lalu digambar dengan **nx.draw()** menggunakan warna simpul biru muda dan sisi abu-abu. Karena graf kosong, hasilnya adalah gambar tanpa elemen yang terlihat.

```python
print("Menambahkan node pada graf")
G.add_node(1)
G.add_node(2)
G.add_nodes_from([3,4,5,6])

nx.set_node_attributes(G, {6: {"color":"red"}})
print(f"Daftar node: {list(G.nodes())}, Jumlah: {G.number_of_nodes()}")

plt.figure(figsize=(5,5))
colors = ["red" if G.nodes[n].get("color", "") == "red" else "lightblue" for n in G.nodes()]
nx.draw(G, with_labels=True, node_color=colors, edge_color="gray")
plt.title("Graf dengan node")
plt.show()
```

Menambahkan simpul (node) ke dalam **graf berarah** menggunakan **NetworkX**. Simpul **1** dan **2** ditambahkan satu per satu, sementara simpul **3, 4, 5,** dan **6** ditambahkan sekaligus dengan **add_nodes_from()**. Simpul **6** diberi atribut warna **"red"** menggunakan **set_node_attributes()**. Daftar simpul dan jumlahnya ditampilkan dengan **print()**. Kemudian, graf digambar menggunakan **nx.draw()**, dengan warna merah untuk simpul **6** dan biru muda untuk lainnya. Hasilnya adalah graf dengan enam simpul yang divisualisasikan menggunakan **Matplotlib**.

```python
print("Menambahkan node pada graf")
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edges_from([(1,4),(2,3),(3,4),(4,5),(5,6),(6,1)])

print(f"Daftar edge: {list(G.edges())}, Jumlah: {G.number_of_edges()}")

plt.figure(figsize=(5,5))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color=colors, edge_color="gray", arrows=True)
plt.title("Graf berarah dengan node")
plt.show()
```

Menambahkan **sisi (edge)** ke dalam **graf berarah** menggunakan **NetworkX**. Sisi antara simpul ditambahkan satu per satu dengan **add_edge()** dan sekaligus dengan **add_edges_from()**. Setelah itu, daftar sisi dan jumlahnya ditampilkan dengan **print()**. Graf kemudian divisualisasikan menggunakan **nx.draw()** dengan tata letak **spring_layout()**, warna simpul sesuai atribut sebelumnya, serta panah pada sisi untuk menunjukkan arah graf. Hasilnya adalah graf berarah dengan enam simpul dan hubungan antar simpul yang divisualisasikan menggunakan **Matplotlib**.

```python
print("Menghapus node dan edge")
G.remove_edge(1,2)
G.remove_node(6)
print(f"Daftar edge setelah hapus: {G.edges()}")
print(f"Daftar node setelah hapus: {G.nodes()}")

plt.figure(figsize=(5,5))
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", arrows=True)
plt.title("Graf setelah menghapus node dan edge")
plt.show()
```

Menghapus **sisi (1,2)** dan **simpul 6** dari **graf berarah** menggunakan **remove_edge()** dan **remove_node()**. Setelah penghapusan, daftar sisi dan simpul diperbarui dan ditampilkan dengan **print()**. Graf kemudian digambar ulang menggunakan **nx.draw()**, dengan tata letak yang sama, warna simpul biru muda, dan panah pada sisi. Hasilnya adalah graf berarah yang telah dimodifikasi, dengan satu simpul dan satu sisi dihapus, divisualisasikan menggunakan **Matplotlib**.

```python
print("Elemen pada graf")
print("Nodes dalam graf: ", G.nodes())
print("Edges dalam graf: ", G.edges())
print("Derajat masuk tiap node: ", G.in_degree())
print("Derajat keluar tiap node: ", G.out_degree())
```

Menampilkan elemen-elemen dalam **graf berarah** menggunakan **NetworkX**. Pertama, daftar **simpul (nodes)** dan **sisi (edges)** dalam graf dicetak menggunakan **G.nodes()** dan **G.edges()**. Selanjutnya, derajat masuk (**in-degree**) dan derajat keluar (**out-degree**) dari setiap simpul ditampilkan menggunakan **G.in_degree()** dan **G.out_degree()**, yang menunjukkan jumlah sisi yang masuk dan keluar dari masing-masing simpul. Informasi ini membantu dalam menganalisis struktur dan hubungan dalam graf.

```python
print("Menambah weight pada edge")
G.add_edge(1, 4, weight=2.5)
G.add_edge(2, 3, weight=1.2)

plt.figure(figsize=(5,5))
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", arrows=True)
edge_labels = nx.get_edge_attributes(G,"weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Graf berarah dengan bobot")
plt.show()
```

Menambahkan **bobot (weight)** pada **sisi (edges)** dalam **graf berarah** menggunakan **NetworkX**. Sisi **(1,4)** diberi bobot **2.5**, dan sisi **(2,3)** diberi bobot **1.2** melalui parameter **weight** dalam **add_edge()**. Graf kemudian divisualisasikan menggunakan **Matplotlib**, dengan posisi tetap menggunakan **spring_layout()**. Selain menggambar simpul dan sisi, bobot sisi juga ditampilkan menggunakan **nx.draw_networkx_edge_labels()**, yang menampilkan nilai bobot pada setiap sisi yang memiliki atribut tersebut. Hasilnya adalah graf berarah dengan bobot yang terlihat pada masing-masing sisi yang diberi nilai.

```python
print("Analisis graf")
print("Jumlah node: ", G.number_of_nodes())
print("Jumlah edge: ", G.number_of_edges())

try:
    if nx.has_path(G, 1, 3):
        shortest_path = nx.shortest_path(G, source=1, target=3, weight="weight")
        print("Jalur terpendek dari 1 ke 3: ", shortest_path)
    else:
        print("Tidak ada jalur antara node 1 dan 3")
except nx.NetworkXNoPath:
    print("Tidak ada jalur yang tersedia")
```

Melakukan **analisis graf** menggunakan **NetworkX**. Pertama, jumlah **simpul (nodes)** dan **sisi (edges)** dalam graf dicetak menggunakan **G.number_of_nodes()** dan **G.number_of_edges()**. Kemudian, program memeriksa apakah terdapat jalur dari simpul **1** ke simpul **3** menggunakan **nx.has_path()**. Jika jalur ada, program menghitung dan mencetak **jalur terpendek** berdasarkan bobot menggunakan **nx.shortest_path()** dengan parameter **weight="weight"**. Jika tidak ada jalur, program menampilkan pesan bahwa tidak ada koneksi antara simpul tersebut. **Try-except** digunakan untuk menangani kemungkinan **kesalahan** jika jalur tidak ditemukan.

```python
cities = [
    ["Oradea", "Zerind", {"w":71}], ["Oradea", "Sibiu", {"w":151}],
    ["Zerind", "Arad", {"w":75}], ["Arad", "Sibiu", {"w":140}],
    ["Arad", "Timisoara", {"w":118}], ["Timisoara", "Lugoj", {"w":111}],
    ["Lugoj", "Mehadia", {"w":70}], ["Mehadia", "Drobeta", {"w":75}],
    ["Drobeta", "Craiova", {"w":120}], ["Craiova", "Pitesti", {"w":138}],
    ["Craiova", "Rimnicu Vilcea", {"w":146}], ["Rimnicu Vilcea", "Sibiu", {"w":80}],
    ["Rimnicu Vilcea", "Pitesti", {"w":97}], ["Sibiu", "Fagaras", {"w":99}],
    ["Fagaras", "Bucharest", {"w":211}], ["Pitesti", "Bucharest", {"w":101}],
    ["Bucharest", "Giurgiu", {"w":90}], ["Bucharest", "Urziceni", {"w":85}],
    ["Urziceni", "Hirsova", {"w":98}], ["Hirsova", "Eforie", {"w":86}],
    ["Urziceni", "Vaslui", {"w":142}], ["Vaslui", "Iasi", {"w":92}],
    ["Iasi", "Neamt", {"w":87}]
]

A = nx.Graph(cities)
a_edges = nx.get_edge_attributes(A, "w")
pos = nx.spring_layout(A, seed=6)

plt.figure(figsize=[10,6])
nx.draw(A, pos, with_labels=True, node_shape="s", node_color="aquamarine", edge_color="gray", font_weight="bold")
nx.draw_networkx_edge_labels(A, pos, edge_labels=a_edges)

plt.title("Graf peta kota-kota besar di Eropa")
plt.show()

shortest_paths = nx.shortest_path(A,weight="w")
for source, paths in shortest_paths.items():
    for target, path in paths.items():
        print(f"{source} -> {target}: {path}")
```

Kode menggunakan **NetworkX** untuk merepresentasikan peta kota-kota besar di Eropa dalam bentuk graf berbobot. Setiap kota dihubungkan dengan kota lain melalui suatu lintasan yang memiliki bobot tertentu, yang merepresentasikan jarak antara dua kota. Graf dibuat menggunakan `nx.Graph(cities)`, di mana setiap hubungan antar kota memiliki atribut bobot "w". Kemudian, tata letak graf diatur menggunakan `nx.spring_layout(A, seed=6)`, dan visualisasi dilakukan dengan `nx.draw()`, menampilkan kota sebagai simpul dan jalan sebagai sisi. Selanjutnya, label bobot sisi ditampilkan menggunakan `nx.draw_networkx_edge_labels()`. Setelah graf divisualisasikan, kode menghitung dan mencetak jalur terpendek antara setiap pasangan kota menggunakan `nx.shortest_path(A, weight="w")` untuk menentukan jalur dengan bobot total terkecil.

```python
jawir = [
    ["Jakarta", "Cirebon", {"w":327}], ["Jakarta", "Bandung", {"w":270}],
    ["Cirebon", "Semarang", {"w":305}], ["Cirebon", "Bandung", {"w":120}], ["Cirebon", "Yogyakarta", {"w":210}],
    ["Bandung", "Yogyakarta", {"w":373}],
    ["Semarang", "Yogyakarta", {"w":109}], ["Semarang", "Surabaya", {"w":369}],
    ["Yogyakarta", "Surakarta", {"w":60}],
    ["Surakarta", "Semarang", {"w":97}], ["Surakarta", "Malang", {"w":370}],
    ["Surabaya", "Malang", {"w":94}]
]

B = nx.Graph(jawir)
b_edges = nx.get_edge_attributes(B, "w")

plt.figure(figsize=[10, 3])
pos = nx.spring_layout(B, seed=16)
nx.draw(B, pos, with_labels=True, node_color="LightGreen", edge_color="gray", font_weight="bold")
nx.draw_networkx_edge_labels(B, pos, edge_labels=b_edges)

plt.title("Graf peta kota-kota besar di Indonesia (pulau Jawa)")
plt.show()

shortest_paths = nx.shortest_path(B,weight="w")
for source, paths in shortest_paths.items():
    for target, path in paths.items():
        print(f"{source} -> {target}: {path}")
```

Kode peta kota-kota besar di Pulau Jawa dalam bentuk **graf berbobot** menggunakan **NetworkX**. Setiap kota dihubungkan dengan kota lain melalui jalan yang memiliki atribut bobot **"w"**, yang merepresentasikan jarak antar kota dalam kilometer. Graf dibuat menggunakan `nx.Graph(jawir)`, lalu divisualisasikan dengan `nx.draw()`, di mana setiap kota ditampilkan sebagai simpul berwarna hijau dan jalan sebagai sisi berwarna abu-abu. Label bobot pada sisi juga ditampilkan menggunakan `nx.draw_networkx_edge_labels()`. Setelah graf divisualisasikan, kode menghitung **jalur terpendek berdasarkan bobot jarak** menggunakan `nx.shortest_path(B, weight="w")`, untuk menemukan rute dengan total jarak paling kecil antara setiap pasangan kota.
