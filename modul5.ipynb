```python
import networkx as nx
import matplotlib.pyplot as plt
import sys
```

```python
def show_graph(G, pos, title='Graph Visualization'):
    plt.figure(figsize=(10, 6))
    nx.draw(
        G, pos, with_labels=True, node_color='red', node_size=2000,
        font_color="white", font_weight="bold", width=2
    )
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='blue', font_size=12)
    plt.title(title)
    plt.margins(0.2)
    plt.show()

def create_graph():
    pos = {
        'A': (0, 3), 'B': (-10, 2), 'C': (10, 2), 'D': (-15, 1), 'E': (-5, 1),
        'F': (5, 1), 'G': (15, 1), 'H': (-18, 0), 'I': (-12, 0), 'J': (-8, 0),
        'K': (-2, 0), 'L': (2, 0), 'M': (8, 0), 'N': (12, 0), 'O': (18, 0)
    }
    edges = [
        ('A', 'B'), ('A', 'C'), ('B', 'E'), ('C', 'F'), ('C', 'G'),
        ('D', 'H'), ('D', 'I'), ('E', 'J'), ('E', 'K'), ('F', 'L'), ('F', 'M'),
        ('G', 'N'), ('G', 'O')
    ]
    G = nx.Graph()
    G.add_edges_from(edges)
    return G, pos

def print_graph_info(G):
    print("\nElemen pada Graf:")
    print(f"Daftar Node: {list(G.nodes())}")
    print(f"Daftar Edge: {list(G.edges())}")
    print(f"Jumlah Node: {G.number_of_nodes()}")
    print(f"Jumlah Edge: {G.number_of_edges()}")

def bfs_traversal(G, source, depth_limit=None):
    traversal_type = "BFS dengan depth " + str(depth_limit) if depth_limit else "BFS"
    print(f"\nUrutan node yang dikunjungi ({traversal_type}):")
    for i, (u, v) in enumerate(nx.bfs_edges(G, source=source, depth_limit=depth_limit), start=1):
        print(f"{i}. {u} - {v}")

def main():
    G, pos = create_graph()
    print_graph_info(G)
    show_graph(G, pos)
    bfs_traversal(G, source='A')
    bfs_traversal(G, source='A', depth_limit=2)

if __name__ == "__main__":
    main()
```

### Fungsi `show_graph(G, pos, title='Graph Visualization')`

Fungsi ini digunakan untuk menampilkan visualisasi graf (graph) menggunakan pustaka `matplotlib` dan `networkx`. Di awal fungsi, `plt.figure(figsize=(10, 6))` menentukan ukuran gambar yang akan ditampilkan, yaitu 10x6 inci. Kemudian, `nx.draw()` menggambar graf `G` dengan posisi node sesuai dengan `pos`. Node akan diberi warna merah, ukuran besar (2000), dan label node ditampilkan dalam warna putih tebal. Selanjutnya, `nx.get_edge_attributes(G, 'weight')` mencoba mengambil atribut bobot dari setiap edge, meskipun dalam graf ini belum diberikan bobot secara eksplisit. Lalu, `nx.draw_networkx_edge_labels()` akan menampilkan label pada setiap edge (jika ada), dengan warna biru dan ukuran font 12. Judul grafik akan ditampilkan dengan `plt.title(title)`, margin diberi ruang menggunakan `plt.margins(0.2)`, dan akhirnya grafik ditampilkan dengan `plt.show()`.

### Fungsi `create_graph()`

Fungsi ini bertugas membuat struktur graf. Pertama, `pos` adalah sebuah dictionary yang menyimpan posisi (koordinat x dan y) dari masing-masing node agar bisa divisualisasikan secara teratur. Lalu, `edges` adalah daftar pasangan node yang saling terhubung, membentuk edge atau sisi pada graf. Setelah itu, `G = nx.Graph()` membuat objek graf kosong tak berarah. Kemudian `G.add_edges_from(edges)` menambahkan semua edge ke dalam graf tersebut. Akhirnya, fungsi ini mengembalikan objek graf `G` dan posisi node `pos`.

### Fungsi `print_graph_info(G)`

Fungsi ini digunakan untuk menampilkan informasi dasar tentang graf. Pertama, ia mencetak teks “Elemen pada Graf:” sebagai judul. Kemudian, `list(G.nodes())` mencetak semua node yang ada di graf, sedangkan `list(G.edges())` mencetak semua edge atau sisi yang menghubungkan node. Selain itu, `G.number_of_nodes()` mencetak jumlah total node, dan `G.number_of_edges()` mencetak jumlah total edge. Ini berguna untuk memahami struktur graf secara umum.

### Fungsi `bfs_traversal(G, source, depth_limit=None)`

Fungsi ini melakukan penelusuran graf menggunakan algoritma Breadth-First Search (BFS), yang berarti menjelajahi graf dari node terdekat ke yang lebih jauh secara bertingkat. Parameter `source` menunjukkan node awal penelusuran, dan `depth_limit` digunakan untuk membatasi kedalaman penelusuran. Jika `depth_limit` diberikan, maka BFS hanya akan berjalan sampai tingkat kedalaman tertentu. Judul hasil penelusuran akan dicetak terlebih dahulu sesuai apakah menggunakan batas kedalaman atau tidak. Kemudian, fungsi `nx.bfs_edges()` akan menghasilkan edge-edge yang dilewati selama BFS. Setiap edge dicetak dengan urutan nomor, sehingga kita tahu jalur penelusuran yang diambil dari node awal.

```python
def create_directed_graph():
    G = nx.DiGraph()
    edges = [
        ('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'),
        ('C', 'E'), ('C', 'F'), ('E', 'D'), ('E', 'F')
    ]
    G.add_edges_from(edges)

    pos = {
        'A': (0, 1), 'B': (1, 2), 'C': (1, 0),
        'D': (2, 3), 'E': (2, 1.5), 'F': (2, 0)
    }

    labels = {node: node for node in G.nodes()}
    labels['A'] = 'A\nStart'
    labels['F'] = 'F\nGoal'

    return G, pos, labels

def show_graph(G, pos, labels, title='Graph Visualization'):
    plt.figure(figsize=(8, 6))
    nx.draw(G, pos,
            with_labels=True,
            labels=labels,
            node_color='red',
            node_size=2000,
            font_color='white',
            font_weight='bold',
            width=2,
            arrows=True)

    # Tampilkan label untuk edge jika ada atribut 'weight'
    edge_labels = nx.get_edge_attributes(G, 'weight')
    if edge_labels:
        nx.draw_networkx_edge_labels(G, pos,
                                     edge_labels=edge_labels,
                                     font_color='blue',
                                     font_size=12)

    plt.title(title)
    plt.margins(0.2)
    plt.show()

def print_graph_info(G):
    print("\nElemen pada Graf:")
    print(f"Daftar Node: {list(G.nodes())}")
    print(f"Daftar Edge: {list(G.edges())}")
    print(f"Jumlah Node: {G.number_of_nodes()}")
    print(f"Jumlah Edge: {G.number_of_edges()}")

def bfs_traversal(G, source, depth_limit=None):
    traversal_type = f"BFS dengan depth {depth_limit}" if depth_limit else "BFS"
    print(f"\nUrutan node yang dikunjungi ({traversal_type}):")
    for i, (u, v) in enumerate(nx.bfs_edges(G, source=source, depth_limit=depth_limit), start=1):
        print(f"{i}. {u} -> {v}")

def main():
    G, pos, labels = create_directed_graph()
    print_graph_info(G)
    show_graph(G, pos, labels, title="Directed Graph")
    bfs_traversal(G, source='A', depth_limit=1)

# Jalankan program utama
if __name__ == "__main__":
    main()
```

### Fungsi `create_directed_graph()`

Fungsi ini digunakan untuk membuat graf berarah (directed graph), yaitu graf yang setiap sisi (edge) memiliki arah tertentu dari satu node ke node lain. Pertama, `G = nx.DiGraph()` membuat objek graf berarah menggunakan `networkx`. Kemudian, daftar pasangan node yang membentuk edge dimasukkan dalam variabel `edges`. Setiap pasangan seperti `('A', 'B')` menunjukkan bahwa ada panah dari node A ke node B. Edge-edge ini dimasukkan ke dalam graf dengan `G.add_edges_from(edges)`. Selanjutnya, `pos` adalah dictionary yang berisi koordinat dua dimensi dari setiap node agar bisa ditampilkan secara visual dengan posisi yang rapi. Lalu, dictionary `labels` dibuat untuk memberi label pada setiap node. Secara default, labelnya sama dengan nama nodenya, tetapi label untuk node 'A' diubah menjadi "A\nStart" dan node 'F' menjadi "F\nGoal", agar menandakan bahwa A adalah titik awal dan F adalah tujuan. Fungsi ini mengembalikan graf `G`, posisi `pos`, dan label node `labels`.

### Fungsi `show_graph(G, pos, labels, title='Graph Visualization')`

Fungsi ini digunakan untuk menampilkan visualisasi graf berarah yang telah dibuat. Pertama, `plt.figure(figsize=(8, 6))` mengatur ukuran tampilan grafik. Setelah itu, jika graf memiliki atribut bobot (`weight`) pada sisi-sisinya, maka label bobot tersebut akan ditampilkan dengan warna biru menggunakan `nx.draw_networkx_edge_labels()`. Judul grafik ditampilkan dengan `plt.title(title)`, margin ditambahkan agar tampilan tidak terlalu padat, dan grafik ditampilkan dengan `plt.show()`.

### Fungsi `print_graph_info(G)`

Fungsi ini digunakan untuk menampilkan informasi dasar mengenai graf yang telah dibuat. Fungsi ini mencetak daftar semua node dalam graf (`G.nodes()`), semua edge atau sisi beserta arah panahnya (`G.edges()`), serta jumlah total node dan edge. Tujuan dari fungsi ini adalah untuk membantu pengguna memahami struktur umum dari graf yang sedang digunakan.

### Fungsi `bfs_traversal(G, source, depth_limit=None)`

Fungsi ini menjalankan algoritma Breadth-First Search (BFS), yaitu metode penelusuran graf secara bertingkat dari node awal (`source`). Penelusuran ini bisa dilakukan secara penuh, atau dibatasi kedalamannya menggunakan `depth_limit`. Jika batas kedalaman diberikan, BFS hanya menelusuri hingga tingkat tertentu. Judul penelusuran dicetak berdasarkan apakah depth limit digunakan atau tidak. Kemudian, fungsi `nx.bfs_edges()` menghasilkan daftar edge yang dilewati selama penelusuran BFS. Setiap edge dicetak dalam format `u -> v`, menunjukkan arah dari node asal ke node tujuan. Angka di depan digunakan untuk menunjukkan urutan edge yang dilalui.

```python
def showGraph():
    try:
        file = "graphs/eropa"
        G = nx.read_edgelist(f"{file}.edgelist")

        for u, v, data in G.edges(data=True):
            if 'weight' in data:
                data['weight'] = int(data['weight'])
        labels = {}
        with open(f"{file}.labels") as f:
            for line in f:
                parts = line.strip().split('\t')
                if len(parts) == 3:
                    x, y, label = parts
                    labels[label] = (float(x), float(y))

        positions = {}
        with open(f"{file}.pos") as f:
            for line in f:
                parts = line.strip().split('\t')
                if len(parts) == 3:
                    node, x, y = parts
                    positions[node] = (float(x), float(y))

        # Draw the graph with specified positions
        plt.figure(figsize=[8,4])
        w = nx.get_edge_attributes(G, "weight")
        nx.draw(G, pos=positions, node_shape="s", node_size=50, linewidths=2, node_color="white",edgecolors="black")
        # Display the labels
        for label, (x, y) in labels.items():
            plt.text(x, y, label, fontsize=10, ha='right', va='center_baseline')
        if w:
            nx.draw_networkx_edge_labels(G, pos=positions, edge_labels=w, font_size=8, font_color='blue')

        plt.show()
    except Exception as e:
        print("Cieeee mau run kode ku yaa, nih makan: ")
        sys.exit(e)

showGraph()
edges = [
    ('Arad', 'Zerind', 75),
    ('Arad', 'Sibiu', 140),
    ('Arad', 'Timisoara', 118),
    ('Zerind', 'Oradea', 71),
    ('Oradea', 'Sibiu', 151),
    ('Sibiu', 'Fagaras', 99),
    ('Sibiu', 'Rimnicu Vilcea', 80),
    ('Timisoara', 'Lugoj', 111),
    ('Lugoj', 'Mehadia', 70),
    ('Mehadia', 'Drobeta', 75),
    ('Drobeta', 'Craiova', 120),
    ('Rimnicu Vilcea', 'Craiova', 146),
    ('Rimnicu Vilcea', 'Pitesti', 97),
    ('Craiova', 'Pitesti', 138),
    ('Fagaras', 'Bucharest', 211),
    ('Pitesti', 'Bucharest', 101),
    ('Bucharest', 'Giurgiu', 90),
    ('Bucharest', 'Urziceni', 85),
    ('Urziceni', 'Hirsova', 98),
    ('Urziceni', 'Vaslui', 142),
    ('Hirsova', 'Eforie', 86),
    ('Vaslui', 'Iasi', 92),
    ('Iasi', 'Neamt', 87)
]
#trololol
G = nx.Graph()
G.add_weighted_edges_from(edges)
source = "Arad"

for u, v, data in G.edges(data=True):
    if 'weight' in data:
        data['weight'] = int(data['weight'])

bfs_edges = list(nx.bfs_edges(G, source=source))
bfs_nodes = [source] + [v for u, v in bfs_edges]

print(f"Urutan node yang dikunjungi dari {source}:")
for i, node in enumerate(bfs_nodes, start=1):
    print(f"{i}. {node}", end=" ")

weights_used = any('weight' in data for _, _, data in G.edges(data=True))
print("\nApakah informasi weight digunakan pada BFS?")
print("Tidak" if not weights_used else "Ya")
```

### Penelusuran Graf dengan BFS

Setelah graf selesai dibuat, program melakukan penelusuran graf menggunakan algoritma BFS (Breadth-First Search), yang mencari node dari node awal ke seluruh tetangga terdekat terlebih dahulu, kemudian ke tetangga dari tetangga tersebut, dan seterusnya. Node awal penelusuran ditetapkan sebagai `"Bandung"`. Fungsi `nx.bfs_edges()` digunakan untuk mendapatkan urutan sisi yang dilewati selama penelusuran BFS. Dari daftar sisi tersebut, disusun urutan node yang dikunjungi dengan menambahkan node asal (`source`) dan semua node tujuan yang ditemukan selama penelusuran.

Urutan node yang dikunjungi dicetak satu per satu dalam format bernomor, sehingga pengguna dapat melihat jalur penelusuran BFS dari kota Bandung. Akhirnya, program juga memeriksa apakah informasi bobot (jarak antar kota) digunakan dalam algoritma BFS. Karena BFS tidak mempertimbangkan bobot, maka output program akan menjawab bahwa informasi bobot tidak digunakan (`"Tidak"`). Hal ini menjadi pengingat bahwa jika ingin mempertimbangkan jarak terpendek, perlu menggunakan algoritma seperti Dijkstra, bukan BFS.
