```python
import networkx as nx
import matplotlib.pyplot as plt
import sys
```

```python
# Fungsi untuk menampilkan graf
def show_graph(G, pos=None, title=''):
    if pos is None:
        pos = nx.spring_layout(G)

    nx.draw(
        G,
        pos,
        with_labels=True,
        node_color='red',
        node_size=2000,
        font_color='white',
        font_weight='bold',
        width=5
    )

    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels,
        font_color='blue',
        font_weight='bold',
        font_size=12
    )

    plt.margins(0.2)
    plt.title(title)
    plt.show()

# Posisi node untuk visualisasi
pos = {
    'A': (0, 3),
    'B': (-10, 2), 'C': (10, 2),
    'D': (-15, 1), 'E': (-5, 1), 'F': (5, 1), 'G': (15, 1),
    'H': (-18, 0), 'I': (-12, 0), 'J': (-8, 0), 'K': (-2, 0),
    'L': (2, 0), 'M': (8, 0), 'N': (12, 0), 'O': (18, 0)
}

# Inisialisasi graf
G = nx.Graph()

# Menambahkan edges dengan bobot
edges = [
    ('A', 'B', 1), ('A', 'C', 3),
    ('B', 'D', 2), ('B', 'E', 1),
    ('C', 'F', 1), ('C', 'G', 2),
    ('D', 'H', 2), ('D', 'I', 3),
    ('E', 'J', 3), ('E', 'K', 2),
    ('F', 'L', 3), ('F', 'M', 2),
    ('G', 'N', 2), ('G', 'O', 3)
]

# Tambahkan edge ke dalam graf
G.add_weighted_edges_from(edges)

# Evaluasi element graf
print("Element pada Graf:")
print("Daftar Node:", G.nodes())
print("Daftar Edge:", G.edges())
print("Jumlah Node:", G.number_of_nodes())
print("Jumlah Edge:", G.number_of_edges())

# Visualisasi graf
show_graph(G, pos=pos, title="Tree UCS")
```

Fungsi `show_graph()` ini digunakan untuk menggambarkan graf menggunakan `networkx` dan `matplotlib`. Fungsi ini menerima tiga parameter: `G` (graf yang akan divisualisasikan), `pos` (posisi node dalam graf), dan `title` (judul graf). Jika posisi node (`pos`) tidak diberikan, maka fungsi ini akan secara otomatis menghasilkan posisi node menggunakan algoritma `spring_layout`, yang mendistribusikan node secara seragam dalam ruang dua dimensi berdasarkan gaya tarik dan tolak.

Graf ditampilkan dengan simpul yang berwarna merah, dengan ukuran node sebesar 2000 dan teks label yang berwarna putih dan dicetak tebal. Tepi (edge) graf digambarkan dengan ketebalan 5. Fungsi `nx.get_edge_attributes()` digunakan untuk mendapatkan bobot dari setiap tepi, dan bobot ini ditampilkan di atas sisi-sisi graf dengan warna teks biru dan ukuran font 12. Fungsi ini kemudian menampilkan visualisasi graf dengan judul yang ditentukan melalui parameter `title`.

Selanjutnya, di bagian kode berikutnya, graf baru diciptakan menggunakan `networkx.Graph()` yang berisi simpul-simpul dan sisi-sisi yang menghubungkan simpul tersebut. Setiap sisi memiliki bobot yang diwakili oleh angka yang ditentukan dalam variabel `edges`. Data `edges` berisi pasangan simpul yang dihubungkan serta bobotnya, misalnya `'A'` terhubung dengan `'B'` dengan bobot 1, dan sebagainya. Semua sisi ini ditambahkan ke dalam graf menggunakan `add_weighted_edges_from()`.

Kemudian, informasi tentang elemen-elemen graf seperti daftar node, daftar edge, jumlah node, dan jumlah edge dicetak menggunakan fungsi `print()`. Terakhir, fungsi `show_graph()` dipanggil untuk menampilkan graf dengan posisi node yang sudah didefinisikan dalam dictionary `pos`, serta judul "Tree UCS" untuk menggambarkan visualisasi graf tersebut.

Fungsi heuristik yang digunakan (`heuristable`) selalu mengembalikan nilai `0` untuk setiap pasangan node, yang berarti algoritma ini berfungsi seperti UCS biasa (tanpa pengaruh fungsi heuristik). UCS sendiri mengandalkan bobot sisi pada graf untuk menemukan jalur dengan biaya terkecil. Dengan menggunakan fungsi `nx.astar_path()` dan `nx.astar_path_length()`, kita dapat menemukan jalur terpendek dari node `'A'` ke node `'M'` serta menghitung total biaya dari jalur tersebut.

Visualisasi graf dilakukan dengan fungsi `show_graph()`, yang menggambar graf dengan posisi node yang telah ditentukan sebelumnya. Jika posisi node tidak diberikan, layout default `spring_layout` dari NetworkX digunakan untuk menyusun node dalam ruang dua dimensi. Node digambarkan dengan warna merah dan ukuran besar, dengan label berwarna putih dan font tebal, sedangkan sisi (edge) antara node ditampilkan dengan arah menggunakan panah. Bobot sisi juga ditampilkan di atas setiap sisi dengan warna font biru dan ukuran font 12. Fungsi ini memungkinkan kita untuk memahami struktur graf secara visual.

Graf tersebut adalah graf terarah (`DiGraph`), yang dibangun dengan menambahkan node dan sisi yang menghubungkan node-node tersebut beserta bobot masing-masing sisi. Bobot sisi mencerminkan biaya atau jarak antar node yang terhubung oleh sisi tersebut. Setelah graf dibangun, kode ini juga mencetak informasi tentang elemen-elemen graf, termasuk daftar node, daftar sisi lengkap dengan bobotnya, serta jumlah total node dan sisi dalam graf. Label khusus juga ditambahkan untuk node `'A'` sebagai titik awal dan node `'F'` sebagai tujuan, untuk membedakan keduanya dalam visualisasi.
