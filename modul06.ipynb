```python
import networkx as nx
import matplotlib.pyplot as plt
import sys
```

```python
def create_directed_graph():
    G = nx.DiGraph()
    edges = [
    ('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'),
    ('C', 'E'), ('C', 'F'), ('E', 'D'), ('E', 'F')
  ]
    G.add_edges_from(edges)

    pos = {
        'A': (0, 1), 'B': (1, 2), 'C': (1, 0),
        'D': (2, 3), 'E': (2, 1.5), 'F': (2, 0)
    }

    labels = {node: node for node in G.nodes()}
    labels['A'] = 'A\nStart'
    labels['F'] = 'F\nGoal'

    return G, pos, labels

def show_graph(G, pos, labels, title='Graph Visualization'):
    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, labels=labels,
        node_color='pink', node_size=2000,
        font_color='black', font_weight='bold',
        width=2, arrows=True)

    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels,font_color='blue', font_size=12)

    plt.title(title)
    plt.margins(0.2)
    plt.show()

def print_graph_info(G):
    print("\nElemen pada Graf:")
    print(f"Daftar Node: {list(G.nodes())}")
    print(f"Daftar Edge: {list(G.edges())}")
    print(f"Jumlah Node: {G.number_of_nodes()}")
    print(f"Jumlah Edge: {G.number_of_edges()}")

def dfs_traversal(G, source, depth_limit=None):
    traversal_type = f"DFS dengan depth {depth_limit}" if depth_limit else "DFS"
    print(f"\nUrutan node yang dikunjungi ((traversal_type)):")
    for i, (u, v) in enumerate(nx.dfs_edges(G, source=source, depth_limit=depth_limit), start=1):
        print(f"{i}. {u} - {v}")

def main():
    G, pos, labels = create_directed_graph()
    print_graph_info(G)
    show_graph(G, pos, labels, title="Directed Graph")

    dfs_traversal(G, source='A')
    dfs_traversal(G, source='A', depth_limit=1)

if __name__ == "__main__":
    main()
```

### Fungsi `create_directed_graph()`

Fungsi ini bertugas untuk membuat sebuah graf berarah (directed graph), yaitu graf di mana setiap sisi (edge) memiliki arah dari satu simpul ke simpul lainnya. Pertama, objek graf berarah `G` dibuat dengan `nx.DiGraph()`. Kemudian, ditambahkan beberapa sisi yang menunjukkan arah hubungan antar simpul, seperti dari 'A' ke 'B', dari 'A' ke 'C', dan seterusnya. Selanjutnya, posisi dari setiap simpul ditentukan dalam dictionary `pos`, di mana masing-masing simpul diberi koordinat untuk keperluan visualisasi graf. Selain itu, dibuat juga dictionary `labels` yang berisi label dari setiap simpul. Simpul 'A' diberi label khusus “A\nStart” dan simpul 'F' diberi label “F\nGoal”, untuk menunjukkan bahwa 'A' adalah titik awal dan 'F' adalah tujuan. Fungsi ini akhirnya mengembalikan tiga data utama, yaitu objek graf, posisi node, dan label-label simpul.

### Fungsi `show_graph(G, pos, labels, title='Graph Visualization')`

Kemudian, graf digambar menggunakan `nx.draw()` dengan beberapa pengaturan visual: warna node menjadi pink, ukuran node besar (2000), font label hitam dan tebal, serta panah diaktifkan karena graf ini adalah graf berarah. Label-label simpul ditampilkan sesuai dengan dictionary `labels` yang telah dibuat sebelumnya. Selanjutnya, jika terdapat atribut `weight` pada edge, maka label bobot akan digambarkan dengan warna biru menggunakan `draw_networkx_edge_labels()`. Setelah semua elemen graf selesai digambar, judul graf ditampilkan dan margin visualisasi sedikit diberi ruang sebelum hasilnya ditampilkan menggunakan `plt.show()`.

Fungsi ini menampilkan informasi dasar tentang graf yang telah dibuat. Informasi yang dicetak meliputi daftar simpul (node), daftar sisi (edge), jumlah total simpul, dan jumlah total sisi. Fungsi ini sangat berguna untuk memahami struktur graf secara tekstual sebelum atau sesudah divisualisasikan.

### Fungsi `dfs_traversal(G, source, depth_limit=None)`

Fungsi ini menjalankan proses penelusuran graf menggunakan algoritma Depth-First Search (DFS). Penelusuran dilakukan dari simpul asal (`source`) yang diberikan, dalam hal ini biasanya 'A'. DFS menjelajahi graf dengan cara menyusuri satu cabang sampai ke ujung sebelum kembali dan mencoba cabang lainnya. Jika parameter `depth_limit` diberikan, maka penelusuran hanya dilakukan hingga kedalaman tertentu. Jenis penelusuran dicetak di awal (misalnya "DFS dengan depth 1"), lalu urutan pasangan simpul (edge) yang dilalui selama DFS ditampilkan satu per satu dengan penomoran.

```python
def show_graph(G, pos, title='Graph Visualization'):
    plt.figure(figsize=(10, 6))
    nx.draw(
        G, pos, with_labels=True, node_color='pink', node_size=2000,
        font_color="black", font_weight="bold", width=2
    )
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='blue', font_size=12)
    plt.title(title)
    plt.margins(0.2)
    plt.show()

def create_graph():
    pos = {
        'A': (0, 3), 'B': (-10, 2), 'C': (10, 2), 'D': (-15, 1), 'E': (-5, 1),
        'F': (5, 1), 'G': (15, 1), 'H': (-18, 0), 'I': (-12, 0), 'J': (-8, 0),
        'K': (-2, 0), 'L': (2, 0), 'M': (8, 0), 'N': (12, 0), 'O': (18, 0)
    }
    edges = [
        ('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G'),
        ('D', 'H'), ('D', 'I'), ('E', 'J'), ('E', 'K'), ('F', 'L'), ('F', 'M'),
        ('G', 'N'), ('G', 'O')
    ]

    G = nx.Graph()
    G.add_edges_from(edges)
    return G, pos

def print_graph_info(G):
    print("\nElement pada Graf:")
    print(f"Daftar Node: {list(G.nodes())}")
    print(f"Daftar Edge: {list(G.edges())}")
    print(f"Jumlah Node: {G.number_of_nodes()}")
    print(f"Jumlah Edge: {G.number_of_edges()}")

def dfs_traversal(G, source, depth_limit=None):
    traversal_type = "DFS dengan depth " + str(depth_limit) if depth_limit else "DFS"
    print(f"\nUrutan node yang dikunjungi ({traversal_type}):")
    for i, (u, v) in enumerate(nx.dfs_edges(G, source=source, depth_limit=depth_limit), start=1):
        print(f"{i}. {u} - {v}")

def main():
    G, pos = create_graph()
    print_graph_info(G)
    show_graph(G, pos)

    dfs_traversal(G, source='A')
    dfs_traversal(G, source='A', depth_limit=2)

if __name__ == "__main__":
  main()
```

### Fungsi `dfs_traversal(G, source, depth_limit=None)`

Fungsi ini menjalankan proses penelusuran menggunakan algoritma Depth-First Search (DFS), dimulai dari node `source` yang ditentukan. Jika parameter `depth_limit` diisi, maka penelusuran akan dibatasi hanya sampai kedalaman tertentu. Tipe penelusuran dicetak terlebih dahulu di terminal, lalu seluruh edge yang dilalui selama proses DFS ditampilkan secara berurutan. Format output adalah pasangan simpul (misalnya `A - B`) dan disertai dengan nomor urut penelusuran.

```python
def showGraph():
    try:
        file = "graphs/eropa"
        G = nx.read_edgelist(f"{file}.edgelist")

        for u, v, data in G.edges(data=True):
            if 'weight' in data:
                data['weight'] = int(data['weight'])
        labels = {}
        with open(f"{file}.labels") as f:
            for line in f:
                parts = line.strip().split('\t')
                if len(parts) == 3:
                    x, y, label = parts
                    labels[label] = (float(x), float(y))

        positions = {}
        with open(f"{file}.pos") as f:
            for line in f:
                parts = line.strip().split('\t')
                if len(parts) == 3:
                    node, x, y = parts
                    positions[node] = (float(x), float(y))

        # Draw the graph with specified positions
        plt.figure(figsize=[8,4])
        w = nx.get_edge_attributes(G, "weight")
        nx.draw(G, pos=positions, node_shape="s", node_size=50, linewidths=2, node_color="white",edgecolors="black")
        # Display the labels
        for label, (x, y) in labels.items():
            plt.text(x, y, label, fontsize=10, ha='right', va='center_baseline')
        if w:
            nx.draw_networkx_edge_labels(G, pos=positions, edge_labels=w, font_size=8, font_color='blue')

        plt.show()
    except Exception as e:
        print("Cieeee mau run kode ku yaa, nih makan: ")
        sys.exit(e)

showGraph()
edges = [
    ('Arad', 'Zerind', 75),
    ('Arad', 'Sibiu', 140),
    ('Arad', 'Timisoara', 118),
    ('Zerind', 'Oradea', 71),
    ('Oradea', 'Sibiu', 151),
    ('Sibiu', 'Fagaras', 99),
    ('Sibiu', 'Rimnicu Vilcea', 80),
    ('Timisoara', 'Lugoj', 111),
    ('Lugoj', 'Mehadia', 70),
    ('Mehadia', 'Drobeta', 75),
    ('Drobeta', 'Craiova', 120),
    ('Rimnicu Vilcea', 'Craiova', 146),
    ('Rimnicu Vilcea', 'Pitesti', 97),
    ('Craiova', 'Pitesti', 138),
    ('Fagaras', 'Bucharest', 211),
    ('Pitesti', 'Bucharest', 101),
    ('Bucharest', 'Giurgiu', 90),
    ('Bucharest', 'Urziceni', 85),
    ('Urziceni', 'Hirsova', 98),
    ('Urziceni', 'Vaslui', 142),
    ('Hirsova', 'Eforie', 86),
    ('Vaslui', 'Iasi', 92),
    ('Iasi', 'Neamt', 87)
]
#trololol
G = nx.Graph()
G.add_weighted_edges_from(edges)
source = "Arad"

for u, v, data in G.edges(data=True):
    if 'weight' in data:
        data['weight'] = int(data['weight'])

bfs_edges = list(nx.dfs_edges(G, source=source))
bfs_nodes = [source] + [v for u, v in bfs_edges]

print(f"Urutan node yang dikunjungi dari {source}:")
for i, node in enumerate(bfs_nodes, start=1):
    print(f"{i}. {node}", end=" ")

weights_used = any('weight' in data for _, _, data in G.edges(data=True))
print("\nApakah informasi weight digunakan pada DFS?")
print("Tidak" if not weights_used else "Ya")
```
