```python
data_mahasiswa = [
    ("Budi", 81.24, 73.66, 95.89, 2023),
    ("Siti", 98.52, 84.86, 88.7, 2024),
    ("Asep", 91.96, 71.03, 79.93, 2022),
    ("Dewi", 87.96, 97.28, 71.91, 2024),
    ("Andi", 74.68, 77.76, 79.33, 2023),
    ("Lestari", 74.68, 89.88, 79.76, 2022),
    ("Joko", 71.74, 79.35, 91.89, 2022),
    ("Rina", 95.99, 85.6, 89.13, 2022),
    ("Fajar", 88.03, 86.4, 96.62, 2024),
    ("Indah", 91.24, 75.55, 84.17, 2023),
    ("Agus", 70.62, 99.09, 73.59, 2022),
    ("Fitri", 99.1, 93.25, 91.4, 2022),
    ("Rudi", 94.97, 98.18, 92.82, 2022),
    ("Maya", 76.37, 96.84, 86.84, 2024),
    ("Hadi", 75.45, 87.94, 93.13, 2024),
    ("Sri", 75.5, 97.66, 84.81, 2023),
    ("Wawan", 79.13, 72.65, 85.68, 2024),
    ("Yuni", 85.74, 75.88, 82.83, 2022),
    ("Bambang", 82.96, 71.36, 70.76, 2023),
    ("Ayu", 78.74, 79.76, 73.24, 2022),
]

group_nilai = {}
for nama, ap, md, dkb, tahun in data_mahasiswa:
    if tahun not in group_nilai:
        group_nilai[tahun] = {
            "AP": [],
            "MD": [],
            "DKB": []
        }
    group_nilai[tahun]["AP"].append(ap)
    group_nilai[tahun]["MD"].append(md)
    group_nilai[tahun]["DKB"].append(dkb)
thn_sorted = sorted(group_nilai.keys())
matkul = ["AP", "MD", "DKB"]
nama_matkul = {
    "AP": "Algoritma Pemrograman",
    "MD": "Matematika Diskrit",
    "DKB": "Dasar Kecerdasan Buatan"
}

stats_per_mk = {mk: {"min": [], "avg": [], "max": []} for mk in matkul}

for mk in matkul:
    for thn in thn_sorted:
        scores = group_nilai[thn][mk]
        min_val = min(scores)
        max_val = max(scores)
        avg_val = sum(scores) / len(scores)
        stats_per_mk[mk]["min"].append(min_val)
        stats_per_mk[mk]["avg"].append(avg_val)
        stats_per_mk[mk]["max"].append(max_val)

fig, axes = plt.subplots(1, 3, figsize=(20, 5), tight_layout=True)
bar_width = 0.25
x_positions = range(len(thn_sorted))

for i, mk in enumerate(matkul):
    ax = axes[i]
    x_min = [x for x in x_positions]
    x_avg = [x + bar_width for x in x_positions]
    x_max = [x + bar_width * 2 for x in x_positions]
    ax.bar(x_min, stats_per_mk[mk]["min"], width=bar_width, label='Min')
    ax.bar(x_avg, stats_per_mk[mk]["avg"], width=bar_width, label='Rerata')
    ax.bar(x_max, stats_per_mk[mk]["max"], width=bar_width, label='Max')
    ax.set_xticks([x + bar_width for x in x_positions])
    ax.set_xticklabels(thn_sorted)
    ax.set_title(nama_matkul[mk])
    ax.set_ylabel('Nilai')
    ax.legend()

plt.show()
```

Kode dimulai dengan mendefinisikan data mahasiswa berupa list tuple yang masing-masing berisi nama, nilai tiga mata kuliah, dan tahun. Selanjutnya, dilakukan pengelompokan data berdasarkan tahun dengan membuat dictionary `group_nilai` yang menyimpan nilai-nilai untuk setiap mata kuliah (AP, MD, DKB) per tahun. Kemudian, daftar tahun disortir dan untuk setiap mata kuliah dihitung nilai minimum, rata-rata, dan maksimum dari data tiap tahun, yang hasilnya disimpan dalam dictionary `stats_per_mk`. Pada bagian visualisasi, dibuat tiga subplot secara horizontal dalam satu figure berukuran 20x5 inci menggunakan `plt.subplots()`. Untuk setiap mata kuliah, data statistik (min, avg, max) diplot dengan diagram batang dengan lebar bar 0.25, dan posisi batang disesuaikan agar tidak tumpang tindih. Masing-masing subplot diberi judul sesuai dengan nama mata kuliah, label sumbu, dan legenda untuk membedakan jenis nilai. Akhirnya, seluruh plot ditampilkan dengan `plt.show()`.

```python
import networkx as nx

cities = [
    ["Oradea", "Zerind", {"w":71}], ["Oradea", "Sibiu", {"w":151}],
    ["Zerind", "Arad", {"w":75}], ["Arad", "Sibiu", {"w":140}],
    ["Arad", "Timisoara", {"w":118}], ["Timisoara", "Lugoj", {"w":111}],
    ["Lugoj", "Mehadia", {"w":70}], ["Mehadia", "Drobeta", {"w":75}],
    ["Drobeta", "Craiova", {"w":120}], ["Craiova", "Pitesti", {"w":138}],
    ["Craiova", "Rimnicu Vilcea", {"w":146}], ["Rimnicu Vilcea", "Sibiu", {"w":80}],
    ["Rimnicu Vilcea", "Pitesti", {"w":97}], ["Sibiu", "Fagaras", {"w":99}],
    ["Fagaras", "Bucharest", {"w":211}], ["Pitesti", "Bucharest", {"w":101}],
    ["Bucharest", "Giurgiu", {"w":90}], ["Bucharest", "Urziceni", {"w":85}],
    ["Urziceni", "Hirsova", {"w":98}], ["Hirsova", "Eforie", {"w":86}],
    ["Urziceni", "Vaslui", {"w":142}], ["Vaslui", "Iasi", {"w":92}],
    ["Iasi", "Neamt", {"w":87}]
]

A = nx.Graph(cities)
a_edges = nx.get_edge_attributes(A, "w")
pos = nx.spring_layout(A, seed=6)

plt.figure(figsize=[10,6])
nx.draw(A, pos, with_labels=True, node_shape="s", node_color="aquamarine", edge_color="gray", font_weight="bold")
nx.draw_networkx_edge_labels(A, pos, edge_labels=a_edges)

plt.title("Graf peta kota-kota besar di Eropa")
plt.show()

shortest_paths = nx.shortest_path(A,weight="w")
for source, paths in shortest_paths.items():
    for target, path in paths.items():
        print(f"{source} -> {target}: {path}")
```

```python
jawir = [
    ["Jakarta", "Cirebon", {"w":327}], ["Jakarta", "Bandung", {"w":270}],
    ["Cirebon", "Semarang", {"w":305}], ["Cirebon", "Bandung", {"w":120}], ["Cirebon", "Yogyakarta", {"w":210}],
    ["Bandung", "Yogyakarta", {"w":373}],
    ["Semarang", "Yogyakarta", {"w":109}], ["Semarang", "Surabaya", {"w":369}],
    ["Yogyakarta", "Surakarta", {"w":60}],
    ["Surakarta", "Semarang", {"w":97}], ["Surakarta", "Malang", {"w":370}],
    ["Surabaya", "Malang", {"w":94}]
]

B = nx.Graph(jawir)
b_edges = nx.get_edge_attributes(B, "w")

plt.figure(figsize=[10, 3])
pos = nx.spring_layout(B, seed=16)
nx.draw(B, pos, with_labels=True, node_color="LightGreen", edge_color="gray", font_weight="bold")
nx.draw_networkx_edge_labels(B, pos, edge_labels=b_edges)

plt.title("Graf peta kota-kota besar di Indonesia (pulau Jawa)")
plt.show()

shortest_paths = nx.shortest_path(B,weight="w")
for source, paths in shortest_paths.items():
    for target, path in paths.items():
        print(f"{source} -> {target}: {path}")
```

Kode di atas menggunakan pustaka **NetworkX** untuk membuat, memvisualisasikan, dan mencari jalur terpendek pada graf yang mewakili peta koneksi antar kota. Pada bagian pertama, data kota di Eropa beserta bobot (jarak) antar kota disusun dalam list, kemudian dijadikan objek graf dengan **nx.Graph()**. Posisi node diatur menggunakan **spring_layout()** dan graf divisualisasikan dengan **nx.draw()** serta label sisi dengan **nx.draw_networkx_edge_labels()**.
